Bu notebook, ham veri, islenmis veri ve model ozetlerinin hizli kesfini yapmak icin olusturuldu.

```python
from pathlib import Path

import pandas as pd

ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
RAW_PATH = ROOT / 'data' / 'raw' / 'btc_usd_1day.parquet'
PROCESSED_PATH = ROOT / 'data' / 'processed' / 'btc_usd_1day_features.parquet'
MODEL_SUMMARY_PATH = ROOT / 'artifacts' / 'metrics' / 'model_summary.csv'
WALK_FORWARD_PATH = ROOT / 'artifacts' / 'metrics' / 'walk_forward_summary.csv'
```

```python
raw_df = pd.read_parquet(RAW_PATH)
processed_df = pd.read_parquet(PROCESSED_PATH)

print('Raw shape:', raw_df.shape)
print('Processed shape:', processed_df.shape)
display(raw_df.head())
display(processed_df[['datetime', 'close', 'future_return_1', 'target']].head())
```

```python
model_summary = pd.read_csv(MODEL_SUMMARY_PATH)
walk_forward_summary = pd.read_csv(WALK_FORWARD_PATH)

display(model_summary[model_summary['split'] == 'test'].sort_values(['symbol', 'model']))
display(walk_forward_summary.sort_values(['asset_group', 'symbol']))
```
